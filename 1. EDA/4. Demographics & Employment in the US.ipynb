Converting the metro codes and country codes to dictionary format

```python
metro_code_dict = metro_code.set_index('Code').to_dict()['MetroArea']
country_code_dict = country_code.set_index('Code').to_dict()['Country']
```

Using map() function to assign values from metro area to corresponding metro codes, and countries from corresponding country of birth codes.

```python
cps['Metro Area'] = cps['MetroAreaCode'].map(metro_code_dict)
cps['Country'] = cps['CountryOfBirthCode'].map(country_code_dict)
```
